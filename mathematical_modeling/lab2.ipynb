### Задание Б: ### 

Дано нелинейное дифференциальное уравнение второго порядка

$$ \ddot{x} - \frac{1}{\sqrt{(x+l)^2+a^2}} + \frac{1}{\sqrt{(x-l)^2+a^2}} = 0 $$

Здесь $ l, a $ - параметры задачи.

Найти периодическое решение дифференциального уравнения методом Линдштедта в окрестности устойчивого частного решения $ x = 0, \dot{x} = 0 $ при $ l = 1, a = 3 $. Для этого следует:

1. Разложить нелинейные функции исследуемого уравнения в ряд по возмущениям $ x, \dot{x} $ в окрестности точки покоя $ x = 0, \dot{x} = 0 $ и удержать члены до третьего порядка включительно. Ввести в уравнения движений малый параметр $ \varepsilon $, используя замену переменных $ x = \varepsilon y, \,\, \dot{x} = \varepsilon \dot{y} $, и новое время $ \tau $ по формуле $ \tau = \omega t $, где $$ \omega = \omega_0 + \varepsilon \omega_1 + \varepsilon^2 \omega_2 + ... $$ Получить в явном виде периодическое решение $ y=y(\tau), \,\, \dot{y} = \dot{y}(\tau) $  задачи Коши $ y(0) = y_0, \,\, \dot{y}(0) = 0 $ для преобразованного дифференциального уравнения с точностью до членов порядка $ \varepsilon^3 $. После этого вернуться к старым переменным $ x, \dot{x} $ и получить приближенное выражение периодических колебаний в виде $ x=x(t), \,\, \dot{x} = \dot{x}(t) $.

2. С помощью графического калькулятора построить две сравнительные фазовые кривые на плоскости переменных $ x, \,\, \dot{x} $, соответствующие аналитическому (приближенному) решению и строгому решению задачи Коши (полученному на основе численного счета). Рассмотреть два интервала изменения времени t : $ t \in [0, \, 10], \,\, t \in [0, 1 / \varepsilon] $. Численные значения параметров и начальных условий таковы: $ l = 1, \,\, a = 3 $.

$$

y(0) = 1, \quad \quad \dot{y}(0) = 0, \quad \quad \varepsilon = 0.01

$$


### Решение ###

Из прошлого задания имеем:

$$ f(x) = - \frac{1}{\sqrt{(x+1)^2+9}} + \frac{1}{\sqrt{(x-1)^2+9}}  $$

Разложим нелинейную функцию исследуемого уравнения в ряд Тейлора, удерживая члены до третьего порядка:

$$ f(x) = f(x^*) + \frac{f'(x^*)}{1}(x - x^*) + \frac{f''(x^*)}{2}(x - x^*)^2 + \frac{f'''(x^*)}{6}(x - x^*)^3 + O((x - x^*)^4) $$

Найдем производные в точке покоя:

$$

f'(x) = \frac{2 x +2}{2 \left(\left(x +1\right)^{2}+9\right)^{\frac{3}{2}}}-\frac{2 x -2}{2 \left(\left(x -1\right)^{2}+9\right)^{\frac{3}{2}}} \quad \Rightarrow \quad f'(x^*) = \frac{\sqrt{10}}{50}  \\[20pt] 


f''(x) = -\frac{3 \left(2 x +2\right)^{2}}{4 \left(\left(x +1\right)^{2}+9\right)^{\frac{5}{2}}}+\frac{1}{\left(\left(x +1\right)^{2}+9\right)^{\frac{3}{2}}}+\frac{3 \left(2 x -2\right)^{2}}{4 \left(\left(x -1\right)^{2}+9\right)^{\frac{5}{2}}}-\frac{1}{\left(\left(x -1\right)^{2}+9\right)^{\frac{3}{2}}} \quad \Rightarrow \quad f''(x^*) = 0 \\[20pt]

f'''(x) = \frac{15 \left(2 x +2\right)^{3}}{8 \left(\left(x +1\right)^{2}+9\right)^{\frac{7}{2}}}-\frac{9 \left(2 x +2\right)}{2 \left(\left(x +1\right)^{2}+9\right)^{\frac{5}{2}}}-\frac{15 \left(2 x -2\right)^{3}}{8 \left(\left(x -1\right)^{2}+9\right)^{\frac{7}{2}}}+\frac{9 x -9}{\left(\left(x -1\right)^{2}+9\right)^{\frac{5}{2}}} \quad \Rightarrow \quad f'''(x^*) = -\frac{3 \sqrt{10}}{200}



$$





Подставим $ x^* = 0 $:

$$
f(x) = \frac{\sqrt{10}}{50} x - \frac{\sqrt{10}}{400} x^3
$$

Запишем уравнение движения:

$$
\ddot{x} + \frac{\sqrt{10}}{50} x - \frac{\sqrt{10}}{400} x^3 = 0
$$

Используем замену $ x = \varepsilon y, \,\, \dot{x} = \varepsilon \dot{y} $ и новое время $ \tau = \omega t $. Тогда:

$$
\frac{d}{dt} = \omega \frac{d}{d \tau}, \quad \frac{d^2}{dt^2} = \omega^2 \frac{d^2}{d \tau^2}
$$

После замен получаем уравнения вида:

$$
\omega^2 \varepsilon \frac{d^2 y(\tau)}{d \tau^2} + \frac{\sqrt{10}}{50} \varepsilon y - \frac{\sqrt{10}}{400} \varepsilon^3 y^3 = 0
$$

Можно сократить на $ \varepsilon $:

$$
\omega^2 \frac{d^2 y(\tau)}{d \tau^2} + \frac{\sqrt{10}}{50} y - \frac{\sqrt{10}}{400} \varepsilon^2 y^3 = 0
$$

Преобразуем уравнению к виду:

$$
\omega^2 y'' + \omega_0^2 y = \varepsilon f(y), \quad \omega_0^2 = \frac{\sqrt{10}}{50}
$$

Будем искать частоту колебаний и решение задачи Коши в виде ряда по малому параметру:

$$ 
\omega = \omega_0 + \varepsilon \omega_1 + \varepsilon^2 \omega_2 + \varepsilon^3 \omega_3, \\
y = y_0(\tau) + \varepsilon y_1(\tau)+ \varepsilon^2 y_2(\tau)+ \varepsilon^3 y_3(\tau)
$$

Подставляем в полученное уравнение:


$$
(\omega_0 + \varepsilon \omega_1 + \varepsilon^2 \omega_2 + \varepsilon^3 \omega_3)^2 (y_0(\tau) + \varepsilon y_1(\tau)+ \varepsilon^2 y_2(\tau)+ \varepsilon^3 y_3(\tau))'' + \\[6pt]
+ \omega_0^2 (y_0(\tau) + \varepsilon y_1(\tau)+ \varepsilon^2 y_2(\tau)+ \varepsilon^3 y_3(\tau))  - \\[6pt]
- \frac{\sqrt{10}}{400} \varepsilon^2 (y_0(\tau) + \varepsilon y_1(\tau)+ \varepsilon^2 y_2(\tau)+ \varepsilon^3 y_3(\tau))^3 = 0
$$


Сгруппируем члены по степеням $ \varepsilon $. Для части без $ \varepsilon $:

$$
\begin{cases}

\omega_0^2 y_0(\tau)'' + \omega_0^2 y_0(\tau) = 0 \\
y_0(0) = a \\
y_0(0)' = 0

\end{cases}
$$

Решением задачи является:
 
$$ 
y_0(\tau) = a \cdot cos (\tau) 
$$


Для части с $ \varepsilon $:

$$
\begin{cases}

2 \omega_0 \omega_1 y_0(\tau)'' + \omega_0^2 y_1(\tau)'' + \omega_0^2 y_1(\tau) = 0 \\
y_1(0) = 0 \\
y_1(0)' = 0

\end{cases}
$$

Первое уравнение системы имеет вид:

$$
\omega_0 \omega_1 (a \cdot cos (\tau))'' + \omega_0^2 y_1(\tau)'' + \omega_0^2 y_1(\tau) = 0, \\[8pt]
y_1(\tau)'' + y_1(\tau) = \frac{\omega_1}{\omega_0} a \cdot cos (\tau)

$$

При решении этого уравнения возникает резонанс, избавимся от него занулив секулярный член: 

$$ 

\omega_1 = 0

$$

Решением задачи является:

$$
y_1(\tau) = 0
$$

Для части с $ \varepsilon^2 $:

$$
\begin{cases}

\omega_{0}^{2} y_2(\tau)''  +2 \omega_{0} \omega_{1} y_1(\tau)'' +\left(2 \omega_{0} \omega_{2}+\omega_{1}^{2}\right) y_0(\tau) '' + \omega_{0}^{2} y_2(\tau) -\frac{\sqrt{10}}{400} y_{0}(\tau)^{3}=  0  \\


y_2(0) = 0 \\
y_2(0)' = 0

\end{cases}
$$

Первое уравнение системы:

$$
y_2(\tau)''\omega_{0}^{2}-2 a \cos \! \left(\tau \right) \omega_{0} \omega_{2}+y_{2}\! \left(\tau \right) \omega_{0}^{2}-\frac{\sqrt{10}\, a^{3} \left(\cos^{3}\left(\tau \right)\right)}{400} = 0 \\[6pt]

y_2(\tau)'' + y_2(\tau) = \frac{2 a \omega_2}{\omega_0} \cos \! \left(\tau \right)   + \frac{\sqrt{10}\, a^{3} \left(\cos^{3}\left(\tau \right)\right)}{400 \omega_0^2} \\[6pt]

y_2(\tau)'' + y_2(\tau) = \frac{\left(3200 a \omega_{0} \omega_{2}+3 \sqrt{10}\, a^{3}\right) \cos \! \left(\tau \right)}{1600 \omega_{0}^{2}}+\frac{\sqrt{10}\, a^{3} \cos \! \left(3 \tau \right)}{1600 \omega_{0}^{2}}

$$

При решении этого уравнения возникает резонанс, избавимся от него занулив секулярный член: 

$$
\omega_2 = -\frac{3 \sqrt{10}\, a^{2}}{3200 \omega_{0}} \\
$$

Первое уравнение принимает вид:

$$

y_2(\tau)'' + y_2(\tau) = \frac{\sqrt{10}\, a^{3} \cos \! \left(3 \tau \right)}{1600 \omega_{0}^{2}}

$$

Решением задачи является: 

$$
y_{2}\! \left(\tau \right) = 
-\frac{\cos \! \left(\tau \right) \left(\sin^{2}\left(\tau \right)\right) \sqrt{10}\, a^{3}}{3200 \omega_{0}^{2}}
$$


Имеем решение вида $ y(\tau) = y_0(\tau) + \varepsilon y_1(\tau) + \varepsilon^2 y_2(\tau) $. Подставим полученный выражения:

$$

y(\tau) = a \cdot cos(\tau) - \varepsilon^2 \frac{\cos \! \left(\tau \right) \sin^{2}\left(\tau \right) \sqrt{10}\, a^{3}}{3200 \omega_{0}^{2}}

$$

Сделаем обратную замену: $ \tau = \omega \cdot t $, где $ \omega = \omega_0 + \varepsilon \omega_1 + \varepsilon^2 \omega_2 $.

$$

\omega = \sqrt{\frac{10^{1/2}}{50}} - \varepsilon^2 \frac{3 \cdot 10^{\frac{1}{4}} a^{2} \sqrt{2}}{640}, \quad  
\tau = \left( \sqrt{\frac{10^{1/2}}{50}} - \varepsilon^2 \frac{3 \cdot 10^{\frac{1}{4}} a^{2} \sqrt{2}}{640} \right) \cdot t \\[6pt]

y(t) = a \cdot cos(\left( \sqrt{\frac{10^{1/2}}{50}} - \varepsilon^2 \frac{3 \cdot 10^{\frac{1}{4}} a^{2} \sqrt{2}}{640} \right) \cdot t) 
- \varepsilon^2 \frac{\cos \! \left(\left( \sqrt{\frac{10^{1/2}}{50}} - \varepsilon^2 \frac{3 \cdot 10^{\frac{1}{4}} a^{2} \sqrt{2}}{640} \right) \cdot t \right) \sin^{2}\left(\left( \sqrt{\frac{10^{1/2}}{50}} - \varepsilon^2 \frac{3 \cdot 10^{\frac{1}{4}} a^{2} \sqrt{2}}{640} \right) \cdot t \right) \sqrt{10}\, a^{3}}{3200 \omega_{0}^{2}}

$$
